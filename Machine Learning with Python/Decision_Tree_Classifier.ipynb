```python
import pandas as pd
```

```python
df = pd.read_csv("/content/titanic.csv")
df.head()
```

```python
new_df = df.drop(['PassengerId','Name','SibSp','Parch','Ticket', 'Cabin','Embarked'], axis = 'columns')
new_df.head()
```

```python
train_class = new_df.drop('Survived', axis = 'columns')
target_class = new_df.Survived
```

```python
train_class.head()
```

```python
target_class
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
train_class.Sex = le.fit_transform(train_class.Sex)
```

```python
train_class.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
train_class.Fare = scaler.fit_transform(train_class[['Fare']])
train_class.Age = scaler.fit_transform(train_class[['Age']])
train_class.head()
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(train_class, target_class)
```

```python
model.score(train_class, target_class)
```

```python
model.predict([[2, 0, 60, 70]])
```
