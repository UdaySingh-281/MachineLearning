```python
import pandas as pd
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv("/content/carprices.csv")
```

```python
df
```

```python
dummies = pd.get_dummies(df["Car Model"])
dummies
```

```python
merged = pd.concat([df,dummies],axis = 'columns')
```

```python
merged
```

```python
merged = merged.drop(['Car Model'], axis = 'columns')
```

```python
merged
```

```python
final = merged.drop(['Audi A5'],axis = 'columns')
final
```

```python
X = final.drop(['Sell Price($)'], axis = 'columns')
X
```

```python
Y = final['Sell Price($)']
Y
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

model = lr.fit(X,Y)

model.score(X,Y)
```

```python
model.predict([[50000, 4, 1, 0]])
```
