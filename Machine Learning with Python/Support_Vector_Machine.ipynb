```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
dir(iris)
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
df['Flower_name'] = df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
df_0 = df[df.target==0]
df_1 = df[df.target==1]
df_2 = df[df.target==2]
```

```python
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.scatter(df_0['sepal length (cm)'], df_0['sepal width (cm)'], color='green', marker='+')
plt.scatter(df_1['sepal length (cm)'], df_1['sepal width (cm)'], color='blue', marker='.')
```

```python
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.scatter(df_0['petal length (cm)'], df_0['petal width (cm)'], color='green', marker='+')
plt.scatter(df_1['petal length (cm)'], df_1['petal width (cm)'], color='blue', marker='.')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['target', 'Flower_name'], axis = 'columns')
X.head()
```

```python
y = df.target
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.svm import SVC
model = SVC(kernel = 'linear')
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
model.score(X_test, y_test)
```

## **Load Digits Dataset**

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.data
```

```python
digits.feature_names[0]
```

```python
df = pd.DataFrame(digits.data, digits.target)
df.head()
```

```python
df['target'] = digits.target
df.head()
```

```python
from sklearn.model_selection import train_test_split
X = df.drop('target', axis='columns')
y = df.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.svm import SVC
model = SVC(kernel = 'linear')
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
